# Data Cleaning

```python
import pandas as pd

# load polution csv file
df = pd.read_csv('../../data/raw/pollution_2000_2021.csv')
df
```

```python
# drop unnecessary columns
df1 = df.drop(columns=['Year', 'Month', 'Day', 'O3 1st Max Value', 'O3 1st Max Hour', 'CO 1st Max Value', 'CO 1st Max Hour', 'SO2 1st Max Value', 'SO2 1st Max Hour', 'NO2 1st Max Value', 'NO2 1st Max Hour'])
df1
```

```python
# load city csv file
city = pd.read_csv('../../data/raw/city_population.csv')
city
```

```python
# merge city dataframe to add population column
df2 = pd.merge(df1, city, on=['State', 'County', 'City'], how='left')
df2
```

```python
# load coordinate csv file
cord = pd.read_csv('../../data/raw/coordinates.csv')
cord
```

```python
# split coordinates into Latitude and Longitude
cord[['Latitude', 'Longitude']] = cord['Coordinate'].str.split(', ', expand=True)
cord.drop(columns=['Coordinate'], inplace=True)

# re-order columns
cord = cord[['Address', 'Latitude', 'Longitude', 'State', 'County', 'City']]
cord
```

```python
# merge coordinate data frame to add latitude and longitude
df3 = pd.merge(df2, cord, on=['Address', 'State', 'County', 'City'], how='left')
df3 = df3[['Date', 'Address', 'Latitude', 'Longitude', 'State', 'County', 'City', 'Population (at 2000)', 'O3 Mean', 'O3 AQI', 'CO Mean', 'CO AQI', 'SO2 Mean', 'SO2 AQI', 'NO2 Mean', 'NO2 AQI']]
df3
```

```python
# export the dataframe into csv file and store it in processed folder
df3.to_csv('../../data/processed/pollution.csv', index=False)
```
